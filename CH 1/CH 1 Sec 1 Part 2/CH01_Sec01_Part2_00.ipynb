Pada pelajaran kali ini kita akan mempelajari persamaan sistem linear yang tidak memiliki solusi. Sistem ini biasanya disebut dengan sistem inkonsisten

##### Contoh 1 :
Carilah solusi x, y, dan z untuk sistem sebagai berikut:

    x  + y + z  = 1
    x  - y - 2z = 2
    2x - z      = 1


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# Mencari nilai x,y, dan z
A = np.array([[1.,1,1,1],
              [1,-1,-2,2],
              [2,0,-1,1]])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,
              r2-r1,
              r3-2*r1])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,
              r2,
              r3-r2])
print(A)
```

```python
# Visualisasi Jawaban
# Mengganti point x dan y pada rentang 0 sampai 5
x = np.linspace(-5.,5,20)
y = np.linspace(-5.,5,20)
X,Y = np.meshgrid(x,y)
Z = 1-X-Y
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,color='b')
Z = X/2 - Y/2 + 1
ax.plot_surface(X,Y,Z,color='r')
Z = 2*X - 1
ax.plot_surface(X,Y,Z,color='g')
plt.grid('on')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
# Tahap ketiga
# Mengganti point x dan y pada rentang -1 sampai 2
x = np.linspace(-1.,2,20)
y = np.linspace(-1.,2,20)
X,Y = np.meshgrid(x,y)
Z = 1-X-Y
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,color='b')
Z = X/2 - Y/2 + 1
ax.plot_surface(X,Y,Z,color='r')
Z = 2*X - 1
ax.plot_surface(X,Y,Z,color='g')
plt.grid('on')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
