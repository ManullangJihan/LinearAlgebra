### Matplotlib scatter command

Pada bagian ini kita akan mempelajari scatter command dari library matplotlib. Fungsi ini sangat bagus untuk memvisualisasi persamaan sistem linear yang memiliki 3 persamaan yang tidak diketahui.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,7]
plt.rcParams.update({'font.size':10})

x = np.arange(-1,2,1)
y = np.arange(-1,2,1)
[X,Y] = np.meshgrid(x,y)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X,Y,'o',color='r',markersize=12,markerfacecolor='r')
plt.grid('on')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(right=2) #xmax is your value
plt.xlim(left=-2) #xmin is your value
plt.ylim(top=2) #ymax is your value
plt.ylim(bottom=-2) #ymin is your value
```

```python
# Contoh Kedua
x = np.arange(-1,1.5,0.5)
y = np.arange(-1,1.5,0.5)
fig = plt.figure()
X,Y = np.meshgrid(x,y)
ax = fig.add_subplot(111)
ax.plot(X,Y,'o',color='r',markersize=12,markerfacecolor='r')
plt.grid('on')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(right=2) #xmax is your value
plt.xlim(left=-2) #xmin is your value
plt.ylim(top=2) #ymax is your value
plt.ylim(bottom=-2) #ymin is your value
```

```python
# Contoh ketiga
x = np.arange(-1,1.5,0.5)
y = np.arange(-1,1.5,0.5)
X,Y = np.meshgrid(x,y)
Z = X+Y
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X,Y,Z,'o',color='r')
```

```python
# Contoh keempat
x = np.arange(-1,1.5,0.5)
y = np.arange(-1,1.5,0.5)
X,Y = np.meshgrid(x,y)
Z = X+Y
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph of z = x+y')
```

```python
# Contoh kelima
x = np.arange(-1,1.5,0.5)
y = np.arange(-1,1.5,0.5)
X,Y = np.meshgrid(x,y)
Z = X+Y
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,color='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of z = x+y')
```
