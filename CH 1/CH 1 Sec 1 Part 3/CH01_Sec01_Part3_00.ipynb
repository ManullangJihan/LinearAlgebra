Pada pelajaran kali ini kita akan mempelajari persamaan sistem linear yang memiliki solusi infinite. kita akan memulai pelajaran ini dengan contoh

Contoh 1 :

         y  - z  = 4
    2x + y  + 2z = 4
    2x + 2y + z  = 8

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

A = np.array([[0,1,-1,4],
              [2,1,2,4],
              [2,2,1,8]])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r3,
              r2,
              r1])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,
              r2-r1,
              r3])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,
              r2,
              r3+r2])
print(A)
```

## Visualizing Our Answer

Kita telah melihat dari hasil diatas bahwa titik (0,4,0), (3/2,3,-1) dan (-3/2,5,1) memenuhi kriteria untuk persamaan sistem kita. karena nilai x pada titik ini adalah 0, 3/2, dan -3/2, dapat kita uji pada 20 nilai pada rentang -2 dan 2 untuk x

```python
# Visualizing Our Equation
x = np.linspace(-2,2,20)
y = np.linspace(-2,6,20)
X,Y = np.meshgrid(x,y)
Z = Y-4
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z,color='blue', alpha=0.3)
Z = 2 - X - Y/2
ax.plot_surface(X,Y,Z,color='red',alpha=0.3)
Z = 8 - 2*X - 2*Y
ax.plot_surface(X,Y,Z,color='green',alpha=0.3)
z = np.linspace(-4/3,4/3)
x = -3*z/2
y = z+4
ax.plot(x,y,z,linewidth=4,color='black')
ax.plot(0,4,0,'o',markersize=12,markerfacecolor='r')
ax.plot(3/2,3,-1,'o',markersize=12,markerfacecolor='r')
ax.plot(-3/2,5,1,'o',markersize=12,markerfacecolor='r')
plt.show()
```
