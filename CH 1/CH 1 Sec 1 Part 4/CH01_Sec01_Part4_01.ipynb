Carilah nilai untuk sistem x, y, dan z dari persamaan sistem linear dibawah ini menggunakan augmented matrix dalam bentuk eselon

    3x  + 6y  + 2z = -13
    x   + 2y  + z  = -5
    -5x - 10y - 2z = 19

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy import Matrix

plt.rcParams['figure.figsize'] = [15,10]

# Cara Manual
A = np.array([[3,6,2,-13],[1,2,1,-5],[-5,-10,-2,19]])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r2,r1,r3])

r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,r2-3*r1,r3+5*r1])

r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,-r2,r3])

r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,r2,r3-3*r2])

r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1-r2,r2,r3])
print(A)
```

```python
# Cara Automatis
A = Matrix([[3,6,2,-13],[1,2,1,-5],[-5,-10,-2,19]])
System, piv = A.rref()
print(System)
```
