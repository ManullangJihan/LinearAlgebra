#### Contoh Kedua

Pada contoh ini sistem kita memiliki 2 persamaan yang tidak diketahui, yaitu :

    x + y = 1
    x - y = 3

dapat dengan mudah dilihat bahwa (x,y) = (2,-1) merupakan solusi dari persamaan ini. dua persamaan ini akan bertemu pada titik (2,-1), karena x = 2, kita akan mencoba untuk memvisualisasi persamaan ini pada rentang [-1,5]

```python
import matplotlib.pyplot as plt
import numpy as np'(2,-1)')

x = np.arange(-1,5,0.1)
y = 1-x
y2 = x-3
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
ax.plot(x,y2)
ax.plot(2,-1,'o',color='r')
plt.grid('on')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Solution of x+y=1 and x-y=3')
plt.text(2,-0.5,'(2,-1)', horizontalalignment='center')
```
